```python
# Import các thư viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Cấu hình hiển thị đẹp hơn
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
df = pd.read_csv("winequality-red.csv")
df.head()
```

```python
# Thông tin cơ bản
df.info()

# Kiểm tra giá trị thiếu
print("\nMissing values in each column:\n", df.isnull().sum())

# Kích thước tập dữ liệu
print("\nDataset shape:", df.shape)
```

```python
# Thống kê mô tả
df.describe().T
```

```python
# Trung bình, trung vị, mode, phương sai, độ lệch chuẩn
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0]
variance_values = df.var()
std_values = df.std()
iqr_values = stats.iqr(df)

print("Trung bình:\n", mean_values)
print("\nTrung vị:\n", median_values)
print("\nMode:\n", mode_values)
print("\nPhương sai:\n", variance_values)
print("\nĐộ lệch chuẩn:\n", std_values)
print("\nIQR:\n", iqr_values)
```

```python
# Biểu đồ phân bố điểm chất lượng
sns.histplot(df['quality'], bins=7, kde=True, color='salmon')
plt.title("Phân bố điểm chất lượng rượu đỏ")
plt.xlabel("Điểm chất lượng (quality)")
plt.ylabel("Tần suất")
plt.show()

```

```python
# Boxplot các đặc tính hóa học
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Biểu đồ Boxplot các biến hóa học của rượu đỏ")
plt.show()
```

```python
# Tính IQR và xác định outliers cho từng biến
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    print(f"{col}: {len(outliers)} giá trị ngoại lai")
```

```python
import pandas as pd
df = pd.read_csv("diabetes.csv")  # hoặc tên file bạn upload
df.head()
```

```python
df.describe().T
```

```python
from scipy import stats

median_vals = df.median()
mode_vals = df.mode().iloc[0]
variance_vals = df.var()
std_vals = df.std()
iqr_vals = stats.iqr(df)

print("Median:\n", median_vals)
print("Mode:\n", mode_vals)
print("Variance:\n", variance_vals)
print("Std dev:\n", std_vals)
print("IQR:\n", iqr_vals)
```

```python
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    print(f"{col}: {len(outliers)} ngoại lai")
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Phân bố các biến trong bộ dữ liệu bệnh tiểu đường', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.title('Biểu đồ Boxplot phát hiện giá trị ngoại lai', fontsize=14)
plt.show()
```
